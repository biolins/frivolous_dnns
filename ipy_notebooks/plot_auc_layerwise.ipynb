```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sys
sys.path.append('../')
import experiments

curves = 9
curve_labels = ['Unregularized 1/4x', 'Unregularized 1x', 'Unregularized 4x', 'All Regularizers 1/4x',
                'All Regularizers 1x', 'All Regularizers 4x', 'Random Labels 1/4x', 'Random Labels 1x',
                'Random Labels 4x']
curve_colors = ['lightblue', 'blue', 'darkblue', 'lightgreen', 'limegreen', 'darkgreen', 'salmon', 'red', 'darkred']
curve_styles = ['-', '-', '-', '-', '-', '-', '-', '-', '-']
range_len = 7
knockout_range = np.linspace(0.0, 1.0, num=range_len)
layers = 4
layer_labels = [1, 2, 3, 4]
crosses = 3

results = np.zeros((curves, layers, 3))
results_test = np.zeros((curves, layers, 3))
targeted_results = np.zeros((curves, layers, 3))
targeted_results_test = np.zeros((curves, layers, 3))
# all are indexed [curve][perturbation_level][avg/lowererr/uppererr]

included_nets = [62, 63, 67, 74, 75, 79, 86, 87, 91]

for ID in included_nets:

    print('Processing ID:', ID)
    opt = experiments.opt[ID]

    curve_idx = 0  # unregularized
    if opt.hyper.augmentation:  # all regularizers
        curve_idx += 3
    if opt.dataset.random_labels:
        curve_idx += 6  # random labels
    if opt.dnn.neuron_multiplier[0] == 1:
        curve_idx += 1
    elif opt.dnn.neuron_multiplier[0] == 4:
        curve_idx += 2

    csv_file = experiments.opt[ID].csv_dir + experiments.opt[ID].name + '_robustness.csv'
    df = pd.read_csv(csv_file)

    ablation_results = np.zeros((crosses, layers, range_len))
    ablation_results_test = np.zeros((crosses, layers, range_len))
    targeted_ablation_results = np.zeros((crosses, layers, range_len))
    targeted_ablation_results_test = np.zeros((crosses, layers, range_len))

    for cross in range(crosses):
        for layer in range(layers):
            for amount in range(range_len):
                ablation_results[cross, layer, amount] = df[df['cross_validation'] == cross][df['evaluation_set'] == 'train'] \
                    [df['perturbation_layer'] == str(layer)][df['perturbation_name'] == 'Activation Knockout'] \
                    [df['perturbation_amount'] == knockout_range[amount]]['performance'].iloc[0]
                ablation_results_test[cross, layer, amount] = df[df['cross_validation'] == cross] \
                    [df['evaluation_set'] == 'test'][df['perturbation_layer'] == str(layer)] \
                    [df['perturbation_name'] == 'Activation Knockout'][df['perturbation_amount'] == knockout_range[amount]] \
                    ['performance'].iloc[0]
                targeted_ablation_results[cross, layer, amount] = df[df['cross_validation'] == cross] \
                    [df['evaluation_set'] == 'train'][df['perturbation_layer'] == str(layer)] \
                    [df['perturbation_name'] == 'Activation Knockout Selected'][df['perturbation_amount'] == knockout_range[amount]] \
                    ['performance'].iloc[0]
                targeted_ablation_results_test[cross, layer, amount] = df[df['cross_validation'] == cross] \
                    [df['evaluation_set'] == 'test'][df['perturbation_layer'] == str(layer)] \
                    [df['perturbation_name'] == 'Activation Knockout Selected'][df['perturbation_amount'] == knockout_range[amount]] \
                    ['performance'].iloc[0]
                    
    ablation_results = np.mean(ablation_results, axis=2)  # collapse along range_len
    ablation_results_test = np.mean(ablation_results_test, axis=2)  # collapse along range_len
    targeted_ablation_results = np.mean(targeted_ablation_results, axis=2)  # collapse along range_len
    targeted_ablation_results_test = np.mean(targeted_ablation_results_test, axis=2)  # collapse along range_len

    results[curve_idx, :, 0] = np.mean(ablation_results, axis=0)
    results[curve_idx, :, 1] = results[curve_idx, :, 0] - np.min(ablation_results, axis=0)
    results[curve_idx, :, 2] = np.max(ablation_results, axis=0) - results[curve_idx, :, 0]
    results_test[curve_idx, :, 0] = np.mean(ablation_results_test, axis=0)
    results_test[curve_idx, :, 1] = results_test[curve_idx, :, 0] - np.min(ablation_results_test, axis=0)
    results_test[curve_idx, :, 2] = np.max(ablation_results_test, axis=0) - results_test[curve_idx, :, 0]
    
    targeted_results[curve_idx, :, 0] = np.mean(targeted_ablation_results, axis=0)
    targeted_results[curve_idx, :, 1] = targeted_results[curve_idx, :, 0] - np.min(targeted_ablation_results, axis=0)
    targeted_results[curve_idx, :, 2] = np.max(targeted_ablation_results, axis=0) - targeted_results[curve_idx, :, 0]
    targeted_results_test[curve_idx, :, 0] = np.mean(targeted_ablation_results_test, axis=0)
    targeted_results_test[curve_idx, :, 1] = targeted_results_test[curve_idx, :, 0] - np.min(targeted_ablation_results_test, axis=0)
    targeted_results_test[curve_idx, :, 2] = np.max(targeted_ablation_results_test, axis=0) - targeted_results_test[curve_idx, :, 0]
```

```python
# first plot train curves

% matplotlib inline
plt.figure(figsize=(10, 7))
plt.title('CIFAR-10 - Area Under Random Ablation Curve: Train Set', fontsize=20)
plt.xlabel('Layer', fontsize=20)
plt.ylabel('AUC Random Ablation', fontsize=20)
plt.xlim((0.98, layers+0.02))
plt.ylim((-0.01, 1.01))
plt.grid(True, which='both', c='black')
plt.xticks(layer_labels, fontsize=15)
plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1], fontsize=15)

for curve_idx in range(curves):
    plt.errorbar(layer_labels, results[curve_idx, :, 0], yerr=results[curve_idx, :, 1:].T,
                 color=curve_colors[curve_idx], linestyle=curve_styles[curve_idx], 
                 linewidth=4, capsize=5, capthick=2, alpha=0.75)
plt.legend(curve_labels, loc='lower right', framealpha=0.8, fontsize=15)

plt.show()

# second plot test curves

plt.figure(figsize=(10, 7))
plt.title('CIFAR-10 - Area Under Random Ablation Curve: Test Set', fontsize=20)
plt.xlabel('Layer', fontsize=20)
plt.ylabel('AUC Random Ablation', fontsize=20)
plt.xlim((0.98, layers+0.02))
plt.ylim((-0.01, 1.01))
plt.grid(True, which='both', c='black')
plt.xticks(layer_labels, fontsize=15)
plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1], fontsize=15)

for curve_idx in range(curves):
    plt.errorbar(layer_labels, results_test[curve_idx, :, 0], yerr=results_test[curve_idx, :, 1:].T,
                 color=curve_colors[curve_idx], linestyle=curve_styles[curve_idx], 
                 linewidth=4, capsize=5, capthick=2, alpha=0.75)
plt.legend(curve_labels, loc='lower right', framealpha=0.8, fontsize=15)

plt.show()

plt.figure(figsize=(10, 7))
plt.title('CIFAR-10 - Area Under Targeted Ablation Curve: Train Set', fontsize=20)
plt.xlabel('Layer', fontsize=20)
plt.ylabel('AUC Targeted Ablation', fontsize=20)
plt.xlim((0.98, layers+0.02))
plt.ylim((-0.01, 1.01))
plt.grid(True, which='both', c='black')
plt.xticks(layer_labels, fontsize=15)
plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1], fontsize=15)

for curve_idx in range(curves):
    plt.errorbar(layer_labels, targeted_results[curve_idx, :, 0], yerr=targeted_results[curve_idx, :, 1:].T,
                 color=curve_colors[curve_idx], linestyle=curve_styles[curve_idx], 
                 linewidth=4, capsize=5, capthick=2, alpha=0.75)
plt.legend(curve_labels, loc='lower right', framealpha=0.8, fontsize=15)

plt.show()

# second plot test curves

plt.figure(figsize=(10, 7))
plt.title('CIFAR-10 - Area Under Targeted Ablation Curve: Test Set', fontsize=20)
plt.xlabel('Layer', fontsize=20)
plt.ylabel('AUC Targeted Ablation', fontsize=20)
plt.xlim((0.98, layers+0.02))
plt.ylim((-0.01, 1.01))
plt.grid(True, which='both', c='black')
plt.xticks(layer_labels, fontsize=15)
plt.yticks([0, 0.2, 0.4, 0.6, 0.8, 1], fontsize=15)

for curve_idx in range(curves):
    plt.errorbar(layer_labels, targeted_results_test[curve_idx, :, 0], yerr=targeted_results_test[curve_idx, :, 1:].T,
                 color=curve_colors[curve_idx], linestyle=curve_styles[curve_idx], 
                 linewidth=4, capsize=5, capthick=2, alpha=0.75)
plt.legend(curve_labels, loc='lower right', framealpha=0.8, fontsize=15)

plt.show()
```
